# Filter aggregations with HAVING

```python
%%sql
SELECT name, river, lake, sea FROM river LIMIT 10;
```

```python
%%sql
SELECT sea, COUNT(name) FROM river GROUP BY sea
```

```python
%%sql
SELECT sea, COUNT(name) FROM river GROUP BY sea HAVING count(name)>10
--having take the results from the group by, filters that where the aggr results needs the criteria
```

table -> where (taking the rows of the table) -> table (filtered rows) -> group by (aggregation) -> aggr table -> having (match the rows of the aggr, filters that table)

```python
%%sql
SELECT * FROM religion LIMIT 10
```

```python
%%sql
SELECT name,count(country) FROM religion GROUP BY name ORDER BY count(country)DESC
```

```python
%%sql
SELECT name,count(country) 
FROM religion 
WHERE percentage <5
GROUP BY name 
HAVING count(country)=1
```

```python
%%sql 
select * from country limit 5;
```

```python
%%sql 
SELECT city.name, city.population, country.code, country.name
FROM city
JOIN country ON city.country=country.code
```

```python
%%sql 
select * from airport limit 5;
```

```python
%%sql
SELECT airport.IATAcode, airport.name, country.name
FROM airport JOIN country ON airport.country = country.code --pull the country table info to airport tbl
--can be from either table joining 
```

```python
%%sql
SELECT airport.IATAcode, airport.name, country.name, country.population
FROM airport JOIN country ON airport.country = country.code 
WHERE country.population <50000000 AND airport.elevation >1000
ORDER BY airport.IATACode
```

```python
%%sql
SELECT *
FROM city JOIN country ON city.country=country.code
```

```python
%%sql
SELECT country.name, sum(city.population)
FROM city JOIN country ON city.country=country.code
GROUP BY country.name
```

```python
%%sql
SELECT country.name, sum(city.population)
FROM city JOIN country ON city.country=country.code
GROUP BY country.name --or city.country, country.code(unique)
HAVING sum(city.population) > 20000000 --only having 20m
ORDER BY sum(city.population) DESC
```
